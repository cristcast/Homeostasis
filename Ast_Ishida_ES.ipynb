## **2. Componentes**

### **2.1. Potencial de membrana**

El potencial de membrana $\psi_L$ depende de las concentraciones de iones en el interior del lumen. Para ello, la diferencia de potencial en el capacitor:

\begin{equation}
\psi_L = \frac{Q}{C} = \frac{F \cdot V}{C_m \cdot S} \bigg([K^+]_L+[Na^+]_L+[H^+]_L+2[Ca^{2+}]_L-[Cl^-]_L - B\bigg) \;\;\;\;\;\;\;\;\;\;(1)
\end{equation}

En general:

\begin{equation}
\Delta \psi_L = \frac{Q}{C} = \frac{F  V}{C_m  S} \bigg(\sum_{i=1}^{n} z_i[C_i]_L - B\bigg) \;\;\;\;[V]\;\;\;\;\;\;(1)
\end{equation}

En donde:

* *$F$* es la constante de Faraday $[mol/C]$
* *$V$* es el volumen del lisosoma $[L]$
* *$Cm$* es la capacitancia específica $[F/cm^2]$
* *$S$* es la superficie de la membrana $[cm^2]$
* *$[C_i]$* es la concentración interna *_i_* de los *_n_* iones presentes en el lisosoma $[M]$
* *$z_i$* es la valencia del ión *_i_* $[adimensional]$
* *$B$* es la concetración de las partículas de Donnan $[M]$

Por convención, el potencial de membrana $\Delta\psi_L$ es **negativo** si el núúmero de aniones dentro del lisosoma (lumen) es relativamente mayor que en el citosol. 

El potencial de membrana total $\Delta\psi_T$ considera las cargas intrínsecas fuera ($\psi_{out}$) y dentro ($\psi_{in}$) de la bicapa de la membrana lisosomal, es decir: 

\begin{equation}
\Delta\psi_T = \Delta\psi_L + (\psi_{out} - \psi_{in}) \;\;\;\;\;\; (2)
\end{equation}



### **2.2. Concentraciones**
Para tomar en cuenta los efectos del potencial de la bicapa lipídica, el pH y las concentraciones iónicas fueron modificados por un factor de Boltzmann.
($i=luminal$ ; $e=citosólica$)

\begin{equation}
[C]_i=[C]_L exp\bigg(\frac{-z\psi_{in}  F}{R T}\bigg) \;\;\;\;\;\;\;\;\;\;\;\; (3)
\end{equation}

\begin{equation}
[C]_e=[C]_C  exp\bigg(\frac{-z \psi_{out} F}{R  T}\bigg) \;\;\;\;\;\;\;\;\;\;\;\; (4)
\end{equation}

En particular, para el pH:


\begin{equation}
pH_i=pH_L + \frac{\psi_{in}  F}{2.3 R  T}\;\;\;\;\;\;\;\;\;\;\;\; (5)
\end{equation}      

\begin{equation}
pH_e=pH_C + \frac{\psi_{out}  F}{2.3 R T}\;\;\;\;\;\;\;\;\;\;\;\; (6)
\end{equation}





### **2.3. Canales pasivos**
Canales de H+, K+, Ca2+ y Na+ que permiten el movimiento pasivo de estos iones. La electrodifusión está determinada por la ecuación de flujo de Goldman–Hodgkin–Katz:

\begin{equation}
J_{[C]}= P_{[C]}   S  \frac{z_C^2U}{1-e^{-z_CU}}  \bigg([C]_i - [C]_e e^{-z_CU}\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (7)
\end{equation}

El flujo pasivo de protones se describe según:
\begin{equation}
J_{H^{+}}= P_{H^{+}} S  \frac{U}{1-e^{-U}} \bigg(10^{-pH_{e}}  e^{-U} - 10^{-pH_{i}}\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (8)
\end{equation}

\begin{equation}
U = \frac{\Delta \psi F} { R T} \;\;\;\;\;\; (9)
\end{equation}

Tal que:
* $P_H$: permeabilidad a H^{+} por unidad de área.
* $U$: potencial de membrana reducido.

El flujo pasivo de los iones restantes se describe de forma similar, de acuerdo a las siguientes ecuaciones:
\begin{equation}
J_{K^{+}}= P_{K^{+}}   S  \frac{U}{1-e^{-U}}  \bigg([K^{+}]_e   e^{-U} - [K^{+}]_i\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (10)
\end{equation}

\begin{equation}
J_{Na^{+}}= P_{Na^{+}}   S \frac{U}{1-e^{-U}}  \bigg([Na^{+}]_e   e^{-U} - [Na^{+}]_i\bigg) \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (11)
\end{equation}

\begin{equation}
J_{Ca^{2+}}= P_{Ca^{2+}}   S \frac{4U}{1-e^{-2U}}  \bigg([Ca{_f}^{2+}]_e   e^{-2U} - [Ca{_f}^{2+}]_i\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (12)
\end{equation}




### **2.4. V-ATPasa**
Bombea protones al lumen lisosomal en contra de su gradiente electroquíímico usando la energía de la hidrólisis de ATP. Su tasa de bombeo depende del pH luminal y del potencial de membrana, como se ve en la ecuación (13): 
\begin{equation}
J_{VATP} = N_{VATP}  J_{VATP1} (pH_L, \Delta \psi)  \;\;\;\;\;\; (13)
\end{equation}

Tal que: 
* $N_{VATP}$: número de V-ATPasas de la membrana lisosomal.
* $J_{VATP1}$: tasa de bombeo de protones de una V-ATPasa bajo diferentes potenciales de membrana y gradientes de pH, de acuerdo al modelo mecanoquímico de Grab *et al* (2001), el cual fue calibrado con datos experimentales de voltaje-corriente. ### **2.5. Intercambiador ClC-7**
Es un intercambiador de Cl/H+ que funciona como contraión de H+, lo cual permite reducir el potencial de membrana del lisosoma y facilitar el bombeo continuo de protones para producir la acidificación del lumen (Ishida, 2013).
Su tasa de bombeo está descrita por la ec.(14):

\begin{equation}
J_{ClC-7} = N_{CLC-7}  A   \Delta\mu_{ClC-7}  \;\;\;\;\;\; (14)
\end{equation}

Tal que:
* $N_{ClC-7}$ es el número de intercambiadores ClC-7 

* $\Delta\mu_{ClC-7}$ es la fuerza de empuje del flujo contraión, dada por ec.(15)

\begin{equation}
\Delta\mu_{ClC-7}= (n_{H^+}^{ClC-7}+n_{Cl-}^{ClC-7}) \Delta\psi+\frac{R  T}{F} \Bigg( 2.3 n_{H^+}^{ClC-7}  (pH_{e}-pH_{i})+ n_{Cl-}^{ClC-7} ln\bigg(\frac{[Cl^-]_{e}}{[Cl^-]_{i}}\bigg) \Bigg) \;\;\;\;\;\; (15)
\end{equation}

en donde $n_{H^+}^{ClC-7}$ y $n_{Cl^-}^{ClC-7}$ son las estequiometrías de los protones y el cloruro, respectivamente.

* $A$ describe la actividad del intercambiador:

\begin{equation}
A=a x + b (1-x)  \Delta\mu_{ClC-7}^2\;\;\;\;\;\; (16)
\end{equation}


con $a=0.3$ y $b=1.5 10^{-5}$ determinado por un ajuste de curva de ec.16 de acuerdo a la data experimental de Ishida et al. (2013);  y $X$ es la función de conmutación de la actividad de la bomba:

\begin{equation}
x=0.5+0.5  tanh\bigg( \frac{\Delta\mu_{ClC-7}+250}{75} \bigg) \;\;\;\;\;\; (17)
\end{equation}

Por otro lado, Astaburuaga et al (2019) establece una dinámica en la actividad del intercambiador. La ec.14 de la tasa de bombeo queda como ec.18:

\begin{equation}
J_{ClC-7} = N_{CLC-7}  A_{eff}   \Delta\mu_{ClC-7}  \;\;\;\;\;\; (18)
\end{equation}

Donde $A_{eff}$ está definido como:

\begin{equation}
\frac{dA_{eff}}{dt}= \frac{1}{\tau}(A-A_{eff}) \;\;\;\;\;\;\;\; (19)
\end{equation}

En donde $\tau$:

\begin{equation}
  \tau =\begin{cases}                     
               \tau_{act}\;\; si \;\; A>A_{eff}\\
               \tau_{desact}\;\; si \;\; A<A_{eff}\\     
            \end{cases}               \;\;\;\;\;\; (20)
\end{equation}

Por simplicidad, se definen que los tiempos de (des)activación son proporcionales:

\begin{equation}
\tau_{desact}=r_{act} \tau_{act}     \;\;\;\;\;\;\;\;\;\;\;\;  (21)
\end{equation}


Finalmente, se establece un tercer caso en donde el intercambiador se simula como un canal iónico con una tasa de flujo dada por ec.22:

\begin{equation}
J_{ClC-7^{unc}}= P_{Cl^{-}}  S  \frac{U}{1-e^{-U}}  \bigg([Cl^{-}]_e  - [Cl^{-}]_i e^{-U}\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (22)
\end{equation}

En las funciones de python se agregó un cuarto caso que considera el escenario en donde el intercambiador no existe, por lo tanto, $J_{ClC-7} = 0$


### **2.6. Intercambiador CAX**
 Es un intercambiador protón/Ca+2 que, al igual que ClC-7, funciona como contraión de $H^+$. Astaburuga et al.(2019) describe la tasa de flujo como:

\begin{equation}
J_{Cax} = N_{CAX}  \Delta\mu_{CAX}  \;\;\;\;\;\; (23)
\end{equation}
donde:


*   $N_{CAX}$ es el número de intercambiadores $Ca^+/H^+$
*   $\Delta\mu_{CAX}$ es la fuerza de empuje del intercambiador, dado por ec.24 (basada en la ec.15 de ClC-7).


\begin{equation}
\Delta\mu_{CAX}= (n_{H^+}^{CAX} - 2 n_{Ca^{2+}}^{CAX}) \Delta\psi+\frac{R  T}{F} \Bigg( 2.3 n_{H^+}^{CAX}  (pH_{e}-pH_{i})+ \frac{n_{Ca^{2+}}^{CAX}}{2} ln\bigg(\frac{[Ca^{2+}]_{e}}{[Ca^{2+}]_{i}}\bigg) \Bigg) \;\;\;\;\;\; (24)
\end{equation}

Donde $n_{H^+}^{CAX}$ y $n_{Ca^{2+}}^{CAX}$ son las estequiometríías de $H^+$ y $Ca^{2+}$, respectivamente.







### **2.7. Flujo de $Ca^{2+}$ en TRPML1**
El flujo de $Ca^{2+}$ en el canal TRPML1 es descrito como:


\begin{equation}
J_{TRPML1}= P_{TRPML1}  S  \frac{2U}{1-e^{-2U}}  ([Ca_{f}^{2+}]_e   e^{-2U} - [Ca_{f}^{+}]_i)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (25)
\end{equation}

En donde la permeabilidad $P_{TRPML1}$ se define como:

\begin{equation}
P_{TRPML1} = 3.88  10^{-9}(y|\Delta\psi|+(1-y)\frac{|\Delta\psi+40|^3}{pH_{L}^{2.2}} \;\;\;\;\;\; (26)
\end{equation}

con $\large y$ igual a:
\begin{equation}
y=0.5-0.5tanh(\Delta\psi+40) \;\;\;\;\;\; (27)
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from matplotlib import cm
from scipy import interpolate
from scipy.integrate import odeint
import matplotlib.gridspec as gridspec
from mpl_toolkits.mplot3d import Axes3D
from time import time
from scipy import signal
import itertools

#-----------
# FUNCIONES
#----------
# FUNCIONES NECESARIAS PARA LOS CÁLCULOS
def area(volumen):   # Cálculo de superficie de esfera [cm^2] a partir del volumen [lts]
    r = np.cbrt((volumen*1000)*0.75/np.pi) # lts to cm3 then radio cm
    area_cell = 4*np.pi*r**2 #[cm^2] 
    return area_cell

def psi_f(z, psi):   # Cálculo de potencial de membrana reducido U [mV]
    if abs(psi) > 0.1:
        gg = z*psi / (1 - np.exp(-z*psi/RTF)) / RTF
    else:            # Casos particulares por expansión con series de Taylor
        gg = 1 + psi*z/(2*RTF) + (psi*z/RTF)**2/12 - (z*psi/RTF)**4/720 + (z*psi/RTF)**6/30240
    return gg 

def volumen_ins(h,k,na,cl,ca):
    mol_in = h + k + na + cl + ca + C*init_V
    return mol_in/OSM_C

def psi_vol_A(pH,h,k,na,cl,ca,OSM):     #Calcula el potencial de membrana [mV], volumen [lts] y superficie [cm^2]
    if OSM is True:
        vol = volumen_ins(h,k,na,cl,ca)
        S = area(vol)
    else:
        vol = init_V
        S = S_i
    #psi = (F/(cap_0*S))*(z_H*h + z_K*k + z_Na*na + z_Cl*cl + z_Ca*ca - vol*B)*1000 #mV
    psi = (F/(cap_0*S))*(vol*10**(-pH) + z_K*k + z_Na*na + z_Cl*cl + z_Ca*ca - vol*B)*1000 #mV
    return psi,vol,S

# FUNCIÓN DEL SISTEMA DE ECUACIONES DIFERENCIALES
def EDO3(yy, t): 
    global CLC_H, CLC_Cl   
    dy = [0, 0, 0, 0, 0, 0, 0]
    mH = yy[0]/NA    #moles
    pH = yy[1]       
    mK = yy[2]/NA    #moles
    mNa = yy[3]/NA   #moles
    mCl = yy[4]/NA   #moles
    mCa_T = yy[5]/NA #moles
    mCa_F = yy[6]/NA #moles
    
    # Cálculo de potencial de membrana, volumen y superficie
    psi, Vol, S = psi_vol_A(pH,mH,mK,mNa,mCl,mCa_T,OSMOSIS)
    
    # Número de partículas a concentraciones luminales[M]
    #H = mH/Vol    #(Numero de moles/numero avogadro)/(Volumen[lts]) -> mol/lts -> [M]
    H = 10**(-pH)
    K = mK/Vol
    Na = mNa/Vol
    Cl = mCl/Vol
    Ca_F = mCa_F/Vol
    Ca_T = mCa_T/Vol
    r = Ca_F/Ca_T
    
    # Concentraciones de superficie citoplasmáticas modificadas
    pH_C0   = pH_C + psi_out/(RTF*2.3)
    K_C0    = K_C  * np.exp(-psi_out/RTF)
    Na_C0   = Na_C * np.exp(-psi_out/RTF)
    Cl_C0   = Cl_C * np.exp(psi_out/RTF)
    Ca_F_C0 = Ca_C * np.exp(-2*psi_out/RTF)

    # Concentraciones de superficie luminal modificadas
    pH_L0   = pH   + psi_in/(RTF*2.3)
    K_L0    = K    * np.exp(-z_K*psi_in/RTF)
    Na_L0   = Na   * np.exp(-z_Na*psi_in/RTF)
    Cl_L0   = Cl   * np.exp(-z_Cl*psi_in/RTF)
    Ca_F_L0 = Ca_F * np.exp(-z_Ca*psi_in/RTF)

    # Cambio en pH
    delta_pH = pH_C0 - pH_L0 
    
    # V-ATPase
    J_VATPASE = VATP_bool * N_VATP * f_VATPpump(pH, psi)[0]

    # ClC-7 contraion {H out, Cl in}
    if CLC7_type == "FAST":
        CLC_mu = (CLC_H + CLC_Cl) * psi   +  RTF * (CLC_H * 2.3 * delta_pH + CLC_Cl * np.log(Cl_C0/Cl_L0))
        x      = 0.5 + 0.5 * np.tanh((CLC_mu + 250)/75)
        A      = 0.3 * x + 1.5e-5 * (1-x) * CLC_mu**2
        J_CLC  = N_CLC * A * CLC_mu
    if CLC7_type == "UNC":
        J_CLC = P_Cl * S * (Cl_C0 - Cl_L0 * np.exp(-psi/RTF)) * psi_f(-z_Cl, psi) * NA/1000
        CLC_Cl = 1
        CLC_H = 0
        
    # CAX contraióón {H out, Ca+2 in}
    J_CAX = N_CAX * ((CAX_H - 2*CAX_Ca) * psi + RTF * (CAX_H * 2.3 * delta_pH + CAX_Ca/2 * np.log(Ca_F_L0/Ca_F_C0)))

    # Canal TRPML1
    if ch_TRPML1_bool == 1:
        y = 0.5 - 0.5 * np.tanh(psi + 40)
        P_trpml1 = trplm1 * (y * abs(psi) + (1-y)*(abs(psi + 40)**3)/(pH_L0**q))
        J_Ca_trpml1 = P_trpml1 * S * (Ca_F_C0*np.exp(-2*psi/RTF)-Ca_F_L0) * psi_f(z_Ca, psi) * NA/1000
    else:
        J_Ca_trpml1 = 0
    
    # Canales pasivos H, K, Na, Ca
    J_H  = ch_h_bool *  P_H  * S * z_H  * (10**(-pH_C0)* np.exp(-psi/RTF)  - 10**(-pH_L0)) * psi_f(z_H, psi) * NA/1000
    J_K  = ch_k_bool *  P_K  * S * z_K  * (K_C0        * np.exp(-psi/RTF)  - K_L0)         * psi_f(z_K, psi) * NA/1000
    J_Na = ch_na_bool * P_Na * S * z_Na * (Na_C0       * np.exp(-psi/RTF)  - Na_L0)        * psi_f(z_Na, psi)* NA/1000
    J_Ca = ch_ca_bool * P_Ca * S * z_Ca * (Ca_F_C0     * np.exp(-2*psi/RTF)- Ca_F_L0)      * psi_f(z_Ca, psi)* NA/1000
    J_Cl = ch_cl_bool * P_Cl * S * (Cl_C0 - Cl_L0 * np.exp(-psi/RTF)) * psi_f(-z_Cl, psi) * NA/1000

    # Diferenciales dependientes de tiempo
    dNHdt   = (J_H + (J_VATPASE) - (ex_clc7_bool * CLC_H * J_CLC) - (ex_cax_bool * CAX_H*J_CAX))
    dpHdt   = (-dNHdt/Vol/NA)/beta_pH
    dNKdt   = (J_K)
    dNNadt  = (J_Na )
    dNCldt  = (ex_clc7_bool * CLC_Cl * J_CLC) + J_Cl
    dNCaTdt = (J_Ca + (ex_cax_bool * CAX_Ca * J_CAX) + J_Ca_trpml1)
    dNCaFdt = (dNCaTdt*r)

    # Retorno de diferenciales
    dy = [dNHdt, dpHdt, dNKdt, dNNadt, dNCldt, dNCaTdt, dNCaFdt]
    return dy

def solve_edo(tiempo_int):
    global y_init, B, C
    # Equilibrio osmótico
    if balance is True:
        mH = H_L*init_V
        mNa = Na_L*init_V
        mCl = Cl_L*init_V
        mCa_T = Ca_T_L*init_V
        mK = K_L*init_V
        B = (-init_psi_total*1e-3/(F/(cap_0*S)) + (z_H*mH + z_K*mK + z_Na*mNa + z_Cl*mCl + z_Ca*mCa_T))/init_V
        C = (OSM_C - OSM_L)
        print("B=",B)
        print("C=",C)
    # Tiempo de integración
    t_start = 0.0
    t_end   = float(tiempo_int)
    N_steps = int(t_end)
    t       = np.linspace(t_start, t_end, N_steps+1)
    
    # Configuración de condiciones iniciales
    init_H     = 10**(-pH_L)
    init_pH    = pH_L
    init_NH    = float(init_H   * init_V*NA)  # Número de iones de H
    init_NK    = float(K_L   * init_V*NA)  # NI K
    init_NNa   = float(Na_L  * init_V*NA)  # NI Na
    init_NCl   = float(Cl_L  * init_V*NA)  # NI Cl
    init_NCa_T = float(Ca_T_L* init_V*NA)  # NI Ca totales
    init_NCa_F = float(Ca_F_L* init_V*NA)  # Ni Ca libres
    # Vector de condiciones iniciales
    y_init = [init_NH, init_pH, init_NK, init_NNa, init_NCl, init_NCa_T, init_NCa_F]
    # RUN EXP
    y = odeint(EDO3, y_init, t, hmax=1e-2)
    return y

def plot_data(H_final, pH_final, K_final, Na_final, Cl_final, Ca_final, CaF_final, psi_final, V_final):
    fig = plt.figure(tight_layout=True, figsize=(12, 12))
    gs = gridspec.GridSpec(5, 2)

    ax = fig.add_subplot(gs[0, 0])
    ax.plot(H_final*1000)
    ax.set_ylabel('$H_L[[mM]$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[0, 1])
    ax.plot(pH_final)
    ax.set_ylabel('$pH_L$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[1, 0])
    ax.plot(K_final*1000)
    ax.set_ylabel('$[K]_L [mM]$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[1, 1])
    ax.plot(Na_final*1000)
    ax.set_ylabel('$[Na+]_L [mM]$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[2, 0])
    ax.plot(Cl_final*1000)
    ax.set_ylabel('$[Cl^-]_L [mM]$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[2, 1])
    ax.plot(Ca_final*1000)
    ax.set_ylabel('$[Ca_T^{2+}]_L [mM]$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[3, 0])
    ax.plot(CaF_final*1000)
    ax.set_ylabel('$[Ca_F^{2+}]_L [mM]$', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[3, 1])
    ax.plot((psi_final+psi_out))
    ax.set_ylabel('$\Delta\psi_L$ mV', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    ax = fig.add_subplot(gs[4, :])
    ax.plot(V_final)
    ax.set_ylabel('Volumen [lts]', fontsize='large')
    ax.set_xlabel('Tiempo[s]', fontsize='large')

    # Valores Finales
    print("   Final pH_L: ",pH_final[-1],)
    print("   Final H_L: ",H_final[-1]*1000, "mM")
    print("   Final [K+]_L: ", K_final[-1]*1000, "mM")
    print("   Final [Na+]_L: ",Na_final[-1]*1000, "mM")
    print("   Final [Cl-]_L: ", Cl_final[-1]*1000, "mM")
    print("   Final [CaF+2]_L: ", CaF_final[-1]*1000, "mM")
    print("   Final [CaT+2]_L: ", Ca_final[-1]*1000, "mM")
    print("   Final psi_L: ",psi_final[-1]+psi_out, "mV") 
    print("   Final Volumen: ", V_final[-1], "lts")
    print("   Aumento del Volumen del lisosoma:", (V_final[-1]-init_V)/init_V*100 , "%")
    print("   Tiempo de ejecución:",total_time, "s")
```

```python
def objective(v):
    global n, Na_L, K_L, Cl_L, Ca_T_L, pH_L, P_H, P_Na, P_K, P_Cl,P_Ca, tiempo_int
    # Carga las condiciones iniciales
    Na_L, K_L, Cl_L, Ca_T_L, pH_L, P_H, P_Na, P_K, P_Cl, P_Ca = v
    if verbose:
        print(n, ")", " Na_L=", Na_L, " K_L=", K_L, " Cl_L=", Cl_L, 
              "  Ca_L=",Ca_T_L, " pH_L=",  pH_L, " P_H=",P_H, " P_Na=", P_Na, " P_K=", P_K,
              " P_Cl=", P_Cl, " P_Ca=", P_Ca)
    try:
        y=solve_edo(tiempo_int)
        H_mol  = y[-1, 0]/NA
        K_mol  = y[-1, 2]/NA
        Na_mol = y[-1, 3]/NA
        Cl_mol = y[-1, 4]/NA
        Ca_mol = y[-1, 5]/NA
        CaF_mol= y[-1, 6]/NA
        pH_final = y[-1, 1]
        psi_final, V_Fin, S_final = psi_vol_A(pH_final,H_mol,K_mol,Na_mol,Cl_mol,Ca_mol,OSMOSIS)
        psi=psi_final
        pH=pH_final
    except:
        psi=np.inf
        pH=np.inf
    value=np.asarray([psi,pH])
    if verbose:
        print("     ->Psi=",psi,"mV      pH=",pH)
    n=n+1
    return value

def in_bounds(point, bounds):
    for d in range(len(bounds)):
        if point[d] < bounds[d, 0] or point[d] > bounds[d, 1]:  # Determina si cada punto está dentro de los bordes
            return False
    return True

def eval_score(value,goal):
    dist = np.linalg.norm(value-goal) #Evalúa qué tan cerca está de la meta
    return dist

def roulette_selection(scores, mu):
    # Calcula las probabilidades de selección para cada individuo
    probs = scores / sum(scores)
    # Selecciona índices para los padres
    selected = np.random.choice(len(scores), size=mu, p=probs, replace=False)
    return selected

# Algoritmo de estrategia de evolución 
def es_comma(objective, bounds, n_iter, mu, lam,goal,mutation):
    best, best_eval = None, 1e+100
    # Calcula el número de hijos por padre
    n_children = int(lam / mu)
    # Crea población inicial
    population = list()
    for _ in range(lam):
        candidate = None
        while candidate is None or not in_bounds(candidate, bounds):
            candidate = bounds[:, 0] + np.random.rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
        population.append(candidate)

    for epoch in range(n_iter):
        print(epoch," generación ----------------------------------")
    # Evalua la solución y la puntúa
        solutions = [objective(c) for c in population]
        scores = [eval_score(c,goal) for c in solutions]    
    # Rankea los puntajes en orden ascendente
        ranks = np.argsort(np.argsort(scores))       
    # Selecciona los padres
        selected = [i for i,_ in enumerate(ranks) if ranks[i] < mu]  # Selección de los mejores padres        
        #selected = roulette_selection(scores, mu)   # Selección de los padres por ruleta      
    # Crea array para los hijos
        children = list()
        if verbose:
            print("Seleccionados:",selected)
        for i in selected:
    # Comprueba si las soluciones son las mejores
            if scores[i] < best_eval:
                best, best_eval,best_solution = population[i], scores[i], solutions[i]
                print('%d, Mejor: f(%s) = %.5f' % (epoch, best, best_eval))
                print(best_solution)   
     # Mantiene los padres
            children.append(population[i])   
    # Cruza
        indexlist = np.arange(mu)
        indexcomb = list(itertools.combinations(indexlist, 2))
        if verbose:
            print("Combinaciones",indexcomb)
        for j in range(lam-mu):
            child = None
            while child is None or not in_bounds(child, bounds):
                child = (children[indexcomb[j][0]] + children[indexcomb[j][1]])/2
                if verbose:
                    print("child=", child)
            children.append(child)

    # Mutación
        index_mut=[np.random.randint(mu, lam-1) for _ in range(int(lam*mutation))]
        if verbose:
                print("Index:",index_mut)
        for k in index_mut:
            children_prev=children[k]
            child = None
            while child is None or not in_bounds(child, bounds):
                child  = children_prev*abs(np.random.randn(len(bounds)))
            children[k] = child
            if verbose:
                print("Mutado de:",children_prev," a:",children[k],"=",child)
     # Cambio de población    
        population = children
    return [best, best_eval, best_solution]
```
