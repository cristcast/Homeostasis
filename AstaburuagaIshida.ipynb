## **2. Componentes**

### **2.1. Potencial de membrana**

El potencial de membrana $\Delta\psi_L$ depende de las concentraciones de iones en el interior del lumen. Para ello, la diferencia de potencial en el capacitor:

\begin{equation}
\Delta \psi_L = \frac{Q}{C} = \frac{F \cdot V}{C_m \cdot S} \bigg([K^+]_L+[Na^+]_L+[H^+]_L+[Ca^{2+}]_L-[Cl^-]_L - B\bigg) \;\;\;\;\;\;\;\;\;\;(1)
\end{equation}

Por convención, el potencial de membrana $\Delta\psi_L$ es **negativo** si el núúmero de aniones dentro del lisosoma (lumen) es relativamente mayor que en el citosol. 

El potencial de membrana total $\Delta\psi_T$ considera las cargas intrínsecas fuera ($\psi_{out}$) y dentro ($\psi_{in}$) de la bicapa de la membrana lisosomal, es decir: 
\begin{equation}
\Delta\psi_T = \Delta\psi_L + (\psi_{out} - \psi_{in}) \;\;\;\;\;\; (2)
\end{equation}



### **2.2. Concentraciones**
Para tomar en cuenta los efectos del potencial de la bicapa lipídica, el pH y las concentraciones iónicas fueron modificados por un factor de Boltzmann.
($i=luminal$ ; $e=citosólica$)

\begin{equation}
[C]_i=[C]_L exp\bigg(\frac{-z\psi_{in}  F}{R T}\bigg) \;\;\;\;\;\;\;\;\;\;\;\; (3)
\end{equation}

\begin{equation}
[C]_e=[C]_C  exp\bigg(\frac{-z \psi_{out} F}{R  T}\bigg) \;\;\;\;\;\;\;\;\;\;\;\; (4)
\end{equation}

En particular, para el pH:


\begin{equation}
pH_i=pH_L + \frac{\psi_{in}  F}{2.3 R  T}\;\;\;\;\;\;\;\;\;\;\;\; (5)
\end{equation}      

\begin{equation}
pH_e=pH_C + \frac{\psi_{out}  F}{2.3 R T}\;\;\;\;\;\;\;\;\;\;\;\; (6)
\end{equation}





### **2.3. Canales pasivos**
Canales de H+, K+, Ca2+ y Na+ que permiten el movimiento pasivo de estos iones. La electrodifusión está determinada por la ecuación de flujo de Goldman–Hodgkin–Katz:

\begin{equation}
J_{[C]}= P_{[C]}   S  \frac{z_C^2U}{1-e^{-z_CU}}  \bigg([C]_i - [C]_e e^{-z_CU}\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (7)
\end{equation}

El flujo pasivo de protones se describe según:
\begin{equation}
J_{H^{+}}= P_{H^{+}} S  \frac{U}{1-e^{-U}} \bigg(10^{-pH_{e}}  e^{-U} - 10^{-pH_{i}}\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (8)
\end{equation}

\begin{equation}
U = \frac{\Delta \psi F} { R T} \;\;\;\;\;\; (9)
\end{equation}

Tal que:
* $P_H$: permeabilidad a H^{+} por unidad de área.
* $U$: potencial de membrana reducido.

El flujo pasivo de los iones restantes se describe de forma similar, de acuerdo a las siguientes ecuaciones:
\begin{equation}
J_{K^{+}}= P_{K^{+}}   S  \frac{U}{1-e^{-U}}  \bigg([K^{+}]_e   e^{-U} - [K^{+}]_i\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (10)
\end{equation}

\begin{equation}
J_{Na^{+}}= P_{Na^{+}}   S \frac{U}{1-e^{-U}}  \bigg([Na^{+}]_e   e^{-U} - [Na^{+}]_i\bigg) \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (11)
\end{equation}

\begin{equation}
J_{Ca^{2+}}= P_{Ca^{2+}}   S \frac{4U}{1-e^{-2U}}  \bigg([Ca{_f}^{2+}]_e   e^{-2U} - [Ca{_f}^{2+}]_i\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (12)
\end{equation}




### **2.4. V-ATPasa**
Bombea protones al lumen lisosomal en contra de su gradiente electroquíímico usando la energía de la hidrólisis de ATP. Su tasa de bombeo depende del pH luminal y del potencial de membrana, como se ve en la ecuación (13): 
\begin{equation}
J_{VATP} = N_{VATP}  J_{VATP1} (pH_L, \Delta \psi)  \;\;\;\;\;\; (13)
\end{equation}

Tal que: 
* $N_{VATP}$: número de V-ATPasas de la membrana lisosomal.
* $J_{VATP1}$: tasa de bombeo de protones de una V-ATPasa bajo diferentes potenciales de membrana y gradientes de pH, de acuerdo al modelo mecanoquímico de Grab *et al* (2001), el cual fue calibrado con datos experimentales de voltaje-corriente. ### **2.5. Intercambiador ClC-7**
Es un intercambiador de Cl/H+ que funciona como contraión de H+, lo cual permite reducir el potencial de membrana del lisosoma y facilitar el bombeo continuo de protones para producir la acidificación del lumen (Ishida, 2013).
Su tasa de bombeo está descrita por la ec.(14):

\begin{equation}
J_{ClC-7} = N_{CLC-7}  A   \Delta\mu_{ClC-7}  \;\;\;\;\;\; (14)
\end{equation}

Tal que:
* $N_{ClC-7}$ es el número de intercambiadores ClC-7 

* $\Delta\mu_{ClC-7}$ es la fuerza de empuje del flujo contraión, dada por ec.(15)

\begin{equation}
\Delta\mu_{ClC-7}= (n_{H^+}^{ClC-7}+n_{Cl-}^{ClC-7}) \Delta\psi+\frac{R  T}{F} \Bigg( 2.3 n_{H^+}^{ClC-7}  (pH_{e}-pH_{i})+ n_{Cl-}^{ClC-7} ln\bigg(\frac{[Cl^-]_{e}}{[Cl^-]_{i}}\bigg) \Bigg) \;\;\;\;\;\; (15)
\end{equation}

en donde $n_{H^+}^{ClC-7}$ y $n_{Cl^-}^{ClC-7}$ son las estequiometrías de los protones y el cloruro, respectivamente.

* $A$ describe la actividad del intercambiador:

\begin{equation}
A=a x + b (1-x)  \Delta\mu_{ClC-7}^2\;\;\;\;\;\; (16)
\end{equation}


con $a=0.3$ y $b=1.5 10^{-5}$ determinado por un ajuste de curva de ec.16 de acuerdo a la data experimental de Ishida et al. (2013);  y $X$ es la función de conmutación de la actividad de la bomba:

\begin{equation}
x=0.5+0.5  tanh\bigg( \frac{\Delta\mu_{ClC-7}+250}{75} \bigg) \;\;\;\;\;\; (17)
\end{equation}

Por otro lado, Astaburuaga et al (2019) establece una dinámica en la actividad del intercambiador. La ec.14 de la tasa de bombeo queda como ec.18:

\begin{equation}
J_{ClC-7} = N_{CLC-7}  A_{eff}   \Delta\mu_{ClC-7}  \;\;\;\;\;\; (18)
\end{equation}

Donde $A_{eff}$ está definido como:

\begin{equation}
\frac{dA_{eff}}{dt}= \frac{1}{\tau}(A-A_{eff}) \;\;\;\;\;\;\;\; (19)
\end{equation}

En donde $\tau$:

\begin{equation}
  \tau =\begin{cases}                     
               \tau_{act}\;\; si \;\; A>A_{eff}\\
               \tau_{desact}\;\; si \;\; A<A_{eff}\\     
            \end{cases}               \;\;\;\;\;\; (20)
\end{equation}

Por simplicidad, se definen que los tiempos de (des)activación son proporcionales:

\begin{equation}
\tau_{desact}=r_{act} \tau_{act}     \;\;\;\;\;\;\;\;\;\;\;\;  (21)
\end{equation}


Finalmente, se establece un tercer caso en donde el intercambiador se simula como un canal iónico con una tasa de flujo dada por ec.22:

\begin{equation}
J_{ClC-7^{unc}}= P_{Cl^{-}}  S  \frac{U}{1-e^{-U}}  \bigg([Cl^{-}]_e  - [Cl^{-}]_i e^{-U}\bigg)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (22)
\end{equation}

En las funciones de python se agregó un cuarto caso que considera el escenario donde el intercambiador no existe.


### **2.6. Intercambiador CAX**
 Es un intercambiador protón/Ca+2 que, al igual que ClC-7, funciona como contraión de $H^+$. Astaburuga et al.(2019) describe la tasa de flujo como:

\begin{equation}
J_{Cax} = N_{CAX}  \Delta\mu_{CAX}  \;\;\;\;\;\; (23)
\end{equation}
donde:


*   $N_{CAX}$ es el número de intercambiadores $Ca^+/H^+$
*   $\Delta\mu_{CAX}$ es la fuerza de empuje del intercambiador, dado por ec.24 (basada en la ec.15 de ClC-7).


\begin{equation}
\Delta\mu_{CAX}= (n_{H^+}^{CAX} - 2 n_{Ca^{2+}}^{CAX}) \Delta\psi+\frac{R  T}{F} \Bigg( 2.3 n_{H^+}^{CAX}  (pH_{e}-pH_{i})+ \frac{n_{Ca^{2+}}^{CAX}}{2} ln\bigg(\frac{[Ca^{2+}]_{e}}{[Ca^{2+}]_{i}}\bigg) \Bigg) \;\;\;\;\;\; (24)
\end{equation}

Donde $n_{H^+}^{CAX}$ y $n_{Ca^{2+}}^{CAX}$ son las estequiometríías de $H^+$ y $Ca^{2+}$, respectivamente.







### **2.7. Flujo de $Ca^{2+}$ en TRPML1**
El flujo de $Ca^{2+}$ en el canal TRPML1 es descrito como:


\begin{equation}
J_{TRPML1}= P_{TRPML1}  S  \frac{2U}{1-e^{-2U}}  ([Ca_{f}^{2+}]_e   e^{-2U} - [Ca_{f}^{+}]_i)  \frac {N_A}{ 10^{3}}   \;\;\;\;\;\; (25)
\end{equation}

En donde la permeabilidad $P_{TRPML1}$ se define como:

\begin{equation}
P_{TRPML1} = 3.88  10^{-9}(y|\Delta\psi|+(1-y)\frac{|\Delta\psi+40|^3}{pH_{L}^{2.2}} \;\;\;\;\;\; (26)
\end{equation}

con y igual a:
\begin{equation}
y=0.5-0.5tanh(\Delta\psi+40) \;\;\;\;\;\; (27)
\end{equation}

* ***tiempo*** determina el tiempo final de iteración [s].



```python
# FUNCIÓN DE SISTEMA DE ECUACIONES DIFERENCIALES
def EDO3(yy, t):
    global balance
    global B
    global balance_osm
    global FI_C
    
    dy = [0, 0, 0, 0, 0, 0, 0, 0]
    NH = yy[0]
    pH = yy[1]
    NK = yy[2]
    NNa = yy[3]
    NCl = yy[4]
    NCa_T = yy[5]
    NCa_F = yy[6]
    psi_l = yy[7]
    Vol = yy[8]

    # Concentraciones luminales[M]
    H = NH/Vol/NA    #(Numero de moles/numero avogadro)/(Volumen[lts]) -> mol/lts -> [M]
    K = NK/Vol/NA
    Na = NNa/Vol/NA
    Cl = NCl/Vol/NA
    Ca_F = NCa_F/Vol/NA
    Ca_T = NCa_T/Vol/NA
    r = Ca_F/Ca_T
    
    if balance is True:
        B =  10**(-pH) + z_K*K + z_Na*Na + z_Cl*Cl + z_Ca*Ca_T
        print(B)
        balance = False
        
        
    def radio(volumen):
        r =np.cbrt((volumen)*0.75/np.pi) #Vol to m^3 then radio [m]
        return r # m

    def area(volumen):
        r=radio(volumen)*100 #cm
        area_cell = 4*np.pi*r**2 #cm^2
        return area_cell
    
    #S=area(Vol)
    
    # Osmosis
    if OSMOSIS is True:
        if balance_osm is True:
            FI_C = fi*(10**(-pH) + K + Na + Cl + Ca_T+B)
            print ("->FI_C: ", FI_C)
            balance_osm = False
        dV=J_W*S*vw*(fi*(10**(-pH) + K + Na + Cl + Ca_T+B)-FI_C)*cm3_lt
    else:
        dV=0
    #H = 10**(-pH)
    # Potencial de membrana
    psi = F/(cap_0*S)*Vol*(H + z_K*K + z_Na*Na + z_Cl*Cl + z_Ca*Ca_T - B)
    dpsi_l = psi-psi_l
    
    # Concentraciones de superficie citoplasmáticas modificadas
    pH_C0 = (pH_C+psi_out/(RTF*2.3))
    K_C0 = K_C*np.exp(-psi_out/RTF)
    Na_C0 = Na_C*np.exp(-psi_out/RTF)
    Cl_C0 = Cl_C*np.exp(psi_out/RTF)
    Ca_F_C0 = Ca_C*np.exp(-2*psi_out/RTF)

    # Concentraciones de superficie luminal modificadas
    pH_L0 = (pH+psi_in/(RTF*2.3))
    K_L0 = K*np.exp(-z_K*psi_in/RTF)
    Na_L0 = Na*np.exp(-z_Na*psi_in/RTF)
    Cl_L0 = Cl*np.exp(-z_Cl*psi_in/RTF)
    Ca_F_L0 = Ca_F*np.exp(-z_Ca*psi_in/RTF)

    # Cambio en pH
    delta_pH = pH_C0-pH_L0

    # Casos particulares de potencial de membrana
    def psi_f(z, psi):
        if abs(psi) > 0.1:
            gg = z*psi / (1 - np.exp(-z*psi/RTF)) / RTF
        # Expansión con series de Taylor
        else:
            gg = 1 + psi*z/(2*RTF) + (psi*z/RTF)**2/12 - (z*psi/RTF)**4/720 + (z*psi/RTF)**6/30240
        return gg   
    
    # V-ATPase
    J_VATPASE = VATP_bool*N_VATP*f_VATPpump(pH, psi)[0]

    # ClC-7 contraion {H out, Cl in}
    CLC_mu = (CLC_H + CLC_Cl)*psi + RTF*(CLC_H*2.3*delta_pH + CLC_Cl*np.log(Cl_C0/Cl_L0))
    x = 0.5 + 0.5*np.tanh((CLC_mu + 250)/75)
    A = 0.3*x + 1.5E-5*(1-x)*CLC_mu**2
    J_CLC = N_CLC*A*CLC_mu
        
    # CAX contraióón {H out, Ca+2 in}
    J_CAX = N_CAX*((CAX_H - 2*CAX_Ca)*psi + RTF*(CAX_H*2.3*delta_pH + CAX_Ca/2*np.log(Ca_F_L0/Ca_F_C0)))

    # Canal TRPML1
    y = 0.5 - 0.5*np.tanh(psi + 40)
    P_trpml1 = ch_TRPML1_bool*trplm1*(y*abs(psi) + (1-y)*(abs(psi/1000 + 40)**3)/(pH**q))
    #print("y= ",y)
    #print("P=",P_trpml1)
    J_Ca_trpml1 = P_trpml1*S*(Ca_F_C0*np.exp(-2*psi/RTF)-Ca_F_L0)*psi_f(2, psi)*NA
   
    # Canales H, K, Na, Ca
    J_H  = ch_h_bool *  P_H  * S * z_H  * (10**(-pH_C0)* np.exp(-psi/RTF)  - 10**(-pH_L0)) * psi_f(z_H, psi) * NA/1000
    J_K  = ch_k_bool *  P_K  * S * z_K  * (K_C0        * np.exp(-psi/RTF)  - K_L0)         * psi_f(z_K, psi) * NA/1000
    J_Na = ch_na_bool * P_Na * S * z_Na * (Na_C0       * np.exp(-psi/RTF)  - Na_L0)        * psi_f(z_Na, psi)* NA/1000
    J_Ca = ch_ca_bool * P_Ca * S * z_Ca * (Ca_F_C0     * np.exp(-2*psi/RTF)- Ca_F_L0)      * psi_f(z_Ca, psi)* NA/1000

    # Diferenciales dependientes de tiempo
    dNHdt = J_H + (J_VATPASE) - (ex_clc7_bool*CLC_H*J_CLC) - (ex_cax_bool * CAX_H*J_CAX)
    dpHdt = (-dNHdt/Vol/NA)/beta_pH
    dNKdt = J_K 
    dNNadt = J_Na 
    dNCldt = ex_clc7_bool * CLC_Cl*J_CLC
    dNCaTdt = J_Ca + (ex_cax_bool * CAX_Ca * J_CAX) + J_Ca_trpml1
    dNCaFdt = dNCaTdt*r

    # Retorno de diferenciales
    dy = [dNHdt, dpHdt, dNKdt, dNNadt, dNCldt, dNCaTdt, dNCaFdt, dpsi_l,dV]
    return dy
```

```python
def solve_edo():
    # Load variables
    global y_init
    init_H = 10**(-init_pH)
    init_NH = float(init_H*init_V*NA)       # Número de iones de H
    init_NK = float(init_K*init_V*NA)       # NI K
    init_NNa = float(init_Na*init_V*NA)     # NI Na
    init_NCl = float(init_Cl*init_V*NA)     # NI Cl
    init_NCa_T = float(init_Ca_T*init_V*NA)  # NI Ca totales
    init_NCa_F = float(init_Ca_F*init_V*NA)  # Ni Ca libres
    init_psi = init_psi_total
    cap_par=F/(cap_0*S)*Vol
    # Vector de condiciones iniciales
    y_init = [init_NH, init_pH, init_NK, init_NNa, init_NCl, init_NCa_T, init_NCa_F, init_psi,init_V]
    # RUN EXP
    y = odeint(EDO3, y_init, t, hmax=1e-2)
    return y
```

# 5. Fijación de potencial de membrana a $\psi_L = 50mV$

Para fijar el potencial de membrana a 50mV se usará la función EDO_fixpsi() descrita a continuación.

```python
def EDO_psifix(yy, t):
    global balance
    global B
    global balance_osm
    global FI_C
    global psi_fix
    
    dy = [0, 0, 0, 0, 0, 0, 0, 0]
    NH = yy[0]
    pH = yy[1]
    NK = yy[2]
    NNa = yy[3]
    NCl = yy[4]
    NCa_T = yy[5]
    NCa_F = yy[6]
    psi_l = yy[7]
    Vol = yy[8]

    # Concentraciones luminales
    H = NH/Vol/NA
    K = NK/Vol/NA
    Na = NNa/Vol/NA
    Cl = NCl/Vol/NA
    Ca_F = NCa_F/Vol/NA
    Ca_T = NCa_T/Vol/NA
    r = Ca_F/Ca_T
    
    if balance is True:
        B =  10**(-pH) + z_K*K + z_Na*Na + z_Cl*Cl + z_Ca*Ca_T
        print(B)
        balance = False
        
        
    def radio(volumen):
        r =np.cbrt((volumen/1000)*0.75/np.pi) #Vol to m^3 then radio [m]
        return r # m

    def area(volumen):
        r=radio(volumen)*100 #cm
        area_cell = 4*np.pi*r**2 #cm^2
        return area_cell
    
    #S=area(Vol)
    
    # Osmosis
    if OSMOSIS is True:
        if balance_osm is True:
            FI_C = fi*(10**(-pH) + K + Na + Cl + Ca_T+B)
            print ("->FI_C: ", FI_C)
            balance_osm = False
        dV=J_W*S*vw*(fi*(10**(-pH) + K + Na + Cl + Ca_T+B)-FI_C)*cm3_lt
    else:
        dV=0
    # Potencial de membrana
    if t < 5000:
        psi = psi_fix
        dpsi_l = psi - psi_l
    else:
        psi = F/(cap_0*S)*Vol*(H + z_K*K + z_Na*Na + z_Cl*Cl + z_Ca*Ca_T - B)
        dpsi_l = psi-psi_l
    
    # Concentraciones de superficie citoplasmáticas modificadas
    pH_C0 = (pH_C+psi_out/(RTF*2.3))
    K_C0 = K_C*np.exp(-psi_out/RTF)
    Na_C0 = Na_C*np.exp(-psi_out/RTF)
    Cl_C0 = Cl_C*np.exp(psi_out/RTF)
    Ca_F_C0 = Ca_C*np.exp(-2*psi_out/RTF)

    # Concentraciones de superficie luminal modificadas
    pH_L0 = (pH+psi_in/(RTF*2.3))
    K_L0 = K*np.exp(-z_K*psi_in/RTF)
    Na_L0 = Na*np.exp(-z_Na*psi_in/RTF)
    Cl_L0 = Cl*np.exp(-z_Cl*psi_in/RTF)
    Ca_F_L0 = Ca_F*np.exp(-z_Ca*psi_in/RTF)

    # Cambio en pH
    delta_pH = pH_C0-pH_L0

    # Casos particulares de potencial de membrana
    def psi_f(z, psi):
        if abs(psi) > 0.1:
            gg = z*psi / (1 - np.exp(-z*psi/RTF)) / RTF
        # Expansión con series de Taylor
        else:
            gg = 1 + psi*z/(2*RTF) + (psi*z/RTF)**2/12 - (z*psi/RTF)**4/720 + (z*psi/RTF)**6/30240
        return gg   
    
    # V-ATPase
    J_VATPASE = VATP_bool*N_VATP*f_VATPpump(pH, psi)[0]

    # ClC-7 contraion {H out, Cl in}
    CLC_mu = (CLC_H + CLC_Cl)*psi + RTF*(CLC_H*2.3*delta_pH + CLC_Cl*np.log(Cl_C0/Cl_L0))
    x = 0.5 + 0.5*np.tanh((CLC_mu + 250)/75)
    A = 0.3*x + 1.5E-5*(1-x)*CLC_mu**2
    J_CLC = N_CLC*A*CLC_mu
        
    # CAX contraióón {H out, Ca+2 in}
    J_CAX = N_CAX*((CAX_H - 2*CAX_Ca)*psi + RTF*(CAX_H*2.3*delta_pH + CAX_Ca/2*np.log(Ca_F_L0/Ca_F_C0)))

    # Canal TRPML1
    y = 0.5 - 0.5*np.tanh(psi + 40)
    P_trpml1 = ch_TRPML1_bool*trplm1*(y*abs(psi) + (1-y)*(abs(psi + 40)**3)/(pH**q))
    #print("y= ",y)
    #print("P=",P_trpml1)
    J_Ca_trpml1 = P_trpml1*S*(Ca_F_C0*np.exp(-2*psi/RTF)-Ca_F_L0)*psi_f(2, psi)*NA/1000
   
    # Canales H, K, Na, Ca
    J_H = ch_h_bool * P_H*S*z_H*(10**(-pH_C0)*np.exp(-psi/RTF)-10**(-pH_L0))*psi_f(z_H, psi)*NA/1000
    J_K = ch_k_bool * P_K*S*z_K*(K_C0*np.exp(-psi/RTF)-K_L0)*psi_f(z_K, psi)*NA/1000
    J_Na = ch_na_bool * P_Na*S*z_Na*(Na_C0*np.exp(-psi/RTF)-Na_L0)*psi_f(z_Na, psi)*NA/1000
    J_Ca = ch_ca_bool * P_Ca*S*z_Ca*(Ca_F_C0*np.exp(-2*psi/RTF)-Ca_F_L0)*psi_f(z_Ca, psi)*NA/1000

    # Diferenciales dependientes de tiempo
    dNHdt = J_H + (J_VATPASE) - (ex_clc7_bool*CLC_H*J_CLC) - (ex_cax_bool * CAX_H*J_CAX)
    dpHdt = (-dNHdt/Vol/NA)/beta_pH
    dNKdt = J_K 
    dNNadt = J_Na 
    dNCldt = ex_clc7_bool * CLC_Cl*J_CLC
    dNCaTdt = J_Ca + (ex_cax_bool * CAX_Ca * J_CAX) + J_Ca_trpml1
    dNCaFdt = dNCaTdt*r

    # Retorno de diferenciales
    dy = [dNHdt, dpHdt, dNKdt, dNNadt, dNCldt, dNCaTdt, dNCaFdt, dpsi_l,dV]
    return dy
```

```python
def solve_edo_psifix():
    # Load variables
    global y_init
    init_H = 10**(-init_pH)
    init_NH = float(init_H*init_V*NA)       # Número de iones de H
    init_NK = float(init_K*init_V*NA)       # NI K
    init_NNa = float(init_Na*init_V*NA)     # NI Na
    init_NCl = float(init_Cl*init_V*NA)     # NI Cl
    init_NCa_T = float(init_Ca_T*init_V*NA)  # NI Ca totales
    init_NCa_F = float(init_Ca_F*init_V*NA)  # Ni Ca libres
    init_psi = init_psi_total
    cap_par=F/(cap_0*S)*Vol
    # Vector de condiciones iniciales
    y_init = [init_NH, init_pH, init_NK, init_NNa, init_NCl, init_NCa_T, init_NCa_F, init_psi,init_V]
    #print(y_init)
    # RUN EXP
    #start_time = time()
    y = odeint(EDO_psifix, y_init, t, hmax=1e-2)
    return y
```
